```python
# [1] 모듈로딩
import pandas as pd
```

```python
# [2] Data 준비
file1 = "test_data.csv"
file2 = "test_data2.csv"
```

```python
# [3] csv => DataFrame으로 저장
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
```

```python
# [4] Data 확인 => 기본정보 
df1.info()
print(f"[index] : {df1.index}")
print(f"[Columns] : {df1.columns}")

df2.info()
print(f"[index] : {df2.index}")
print(f"[Columns] : {df2.columns}")
```

```python
# [4] 우리가 만든 함수를 로딩하여 정보 출력  ==> 정보출력
import df_util as utill
```

```python
utill.checkDataFrame(df1, "df1")
utill.checkDataFrame(df2, "df2")
df1
```

```python
# [5] 행 인덱스 설정
# [5-1] df1 첫번째 컬럼을 행 인덱스로 설정 ==> set_index([칼럼 1]) or set_index(칼럼 1)
df11 = df1.set_index(["index"]) # 멀티인덱스
df11.index.name=None
df11.index
print(df11)
# "A" 행 데이터 추출
df11.loc["A"]
```

```python
# "A", "C" 추출
df11.loc[["A", "C"]]
```

```python
# [5-1] df1 첫번째, 두번째 컬럼을 행 인덱스로 설정 ==> set_index([칼럼 1, 칼럼 2])
df12 = df1.set_index(["index", "col1"]) # 멀티인덱스
df12.index.name=None
df12.index
print(df12)

# ('A', 8) 행 추출
df12.loc[('A', 8)]
```

```python
# ('A', 8), ("A", 1) 행 추출
df12.loc[[('A', 8), ("A", 1)]]
```

```python
# label : 멀티인덱스에서 인덱스끼리 구분하기 
# 멀티인덱스 속성들 확인
mIdx = df12.index
print(f'nlevels : {mIdx.nlevels}')  # 멀티 인덱스의 수 
print(f'levels : {mIdx.levels}') # level 0 = ['A', 'B', 'C'], level1 = [1, 2, 3, 4, 5, 8] 
print(f'names : {mIdx.names}')

# 멀티인덱스에서 하나의 level의 이름 변경
mIdx.set_names("a", level = 0)
```

```python
# 멀티인덱스에서 level = 0이 동일한 행 추출
# "A"
df12.loc["A"] # level = 0이 A인 것들
```

```python
# 멀티인덱스에서 level1이 동일한 행 추출
# "8"
# df12.loc[level=1, 8]  ==> ERROR 
df12.xs(8, level = 1)

# xs 메서드: 하위 레벨에 있는 데이터 추출 인덱스
# 변수명.xs(찾을 데이터, 레벨의 위치)
```

```python
# 멀티인덱스에서 행 추가
# - 인덱스의 행추가 => 변수명.loc[새로운 행 인덱스] = 값
df12.loc[("D", 9), :] = [7, 5] # 새로운 행과 값이 추가됨
```
