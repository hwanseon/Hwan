### DataFrame에서 행/열 추가
- 행 방향 추가
- 열 방향 추가

```python
# 모듈로딩
import pandas as pd
```

```python
# 데이터 준비
data = {"name":["홍길동", "마징가", "원더우먼"],
        'job' : ["의적", "로봇", "영웅"],
        'gender' : ["남", "남", "여"],
        "age" : [100, 200, 300],
        "height" : [180, 300, 200]}
```

```python
# DF로 데이터 저장
dataDF = pd.DataFrame(data)
dataDF
```

- [실습] 행 추가 : 변수명.loc["추가할 행 이름"] = 데이터 
    - 변수명.iloc[ ] : 사용 X
    - data의 수 == columns의 수 경우 행 추가 가능 
    - data의 수 == 1인 경우, 행 추가 가능 
    - 그 외의 경우 행 추가 불가능 !!

```python
# 1개 행 추가 => data 수가 열의 개수와 동일한 경우
dataDF.loc[3] = ["베트맨", "박쥐", "남", 40, 90]  # loc에는 숫자 들어와도 상관 X
```

```python
dataDF 
```

```python
dataDF.loc[4] = ["베트맨", "박쥐", "남", 40, 90] 
dataDF
```

```python
## 1개 행 추가 => data 수를 1개로  ==> 모든 칼럼에 동일한 값 저장됨
dataDF.loc[5] = 100
dataDF
```

```python
## 1개 행 추가 => data 수를 2개로 ==> ERROR, 칼럼수랑 똑같은 수의 데이터를 주던가 1개를 주던가 해야함
# dataDF.loc[6] = ["스파이더맨", "거미"]
# dataDF
```

- 열 추가 : 변수명["추가할 열 이름"] = 데이터
    - data의 수 = 행 수 ==> 열 추가 가능
    - data의 수 == 1 ==> 열 추가 가능
    - 그 외의 경우, 열 추가 X

```python
dataDF # 현재 dataframe
```

```python
## 칼럼/열 추가 => weight 추가 // 행 개수 = data 개수
dataDF["weight"] = [90, 88, 83, 78, 75, 73]
dataDF
```

```python
## 칼럼/열 추가 => weight 추가 // 1개 = data 개수 ==> 하나의 값으로 다 채워짐
dataDF["weight2"] = 75
dataDF
```
