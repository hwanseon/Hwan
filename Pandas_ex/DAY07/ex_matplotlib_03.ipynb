##### [Matplotlib 활용 데이터 시각화]
- x, y축 다양한 처리

```python
# [1] 모듈로딩
from matplotlib import pyplot as plt
```

```python
# [2] Data 준비 => Graph 그리기 - 기본
y_data = [1, 3, 5, 7, 9]
x_data = ["2024-01-01", "2024-02-01", "2024-03-01", "2024-04-01", "2024-05-01"]
plt.plot(x_data, y_data, marker = "o")
plt.show()
```

```python
# [2] Data 준비 => Graph 그리기 - x축, y축 범위 지정
y_data = [1, 3, 5, 7, 9]
x_data = ["2024-01-01", "2024-02-01", "2024-03-01", "2024-04-01", "2024-05-01"]
plt.plot(x_data, y_data, marker = "o")
plt.ylim(bottom = 0, top = 15)
plt.show()
```

```python
# [2] Data 준비 => Graph 그리기 - x축, y축 눈금 (tick) 설정
y_data = [1, 3, 5, 7, 9]
x_data = ["2024-01-01", "2024-02-01", "2024-03-01", "2024-04-01", "2024-05-01"]
plt.plot(x_data, y_data, marker = "o")
plt.ylim(bottom = 0, top = 15)
plt.xticks(rotation=45)
plt.show()
```

```python
# [2] Data 준비 => Graph 그리기 - Graph 출력 창 즉 , figure 크기 조정 : plt.figure(figsize = (가로 인치, 세로 인치)) 
y_data = [1, 3, 5, 7, 9]
x_data = ["2024-01-01", "2024-02-01", "2024-03-01", "2024-04-01", "2024-05-01"]

plt.figure(figsize = (4, 2)) # 원하는 창의 크기 설정
plt.plot(x_data, y_data, marker = "o")
plt.ylim(bottom = 0, top = 15)
plt.xticks(rotation=45)
plt.show()
```

```python
# [2] Data 준비 => Graph 그리기 - Graph 출력 창 즉 , figure 자동 맞춤 : tight_layout() 메서드
y_data = [1, 3, 5, 7, 9]
x_data = ["2024-01-01", "2024-02-01", "2024-03-01", "2024-04-01", "2024-05-01"]

plt.plot(x_data, y_data, marker = "o")
plt.ylim(bottom = 0, top = 15)
plt.xticks(rotation=45)
plt.tight_layout()   # 모든 내용이 출력되도록 자동 맞춤 설정
plt.show()
```

```python
# [3] Graph 그리기 - 눈금(ticks) 설정
y_data = [1, 3, 5, 7, 9]
x_data = ["2024-01-01", "2024-02-01", "2024-03-01", "2024-04-01", "2024-05-01"]

plt.plot(x_data, y_data, marker = "o")
plt.ylim(bottom = 0, top = 10)
plt.xlim(left = 0, right = 6)
plt.xticks([0, 2, 4, 6], ["A", "B", "C", "D"], fontsize = 10, fontweight = "bold")
plt.show()
```

```python
# [3] Graph 그리기 - 눈금(ticks) 설정
y_data = [1, 3, 5, 7, 9]
x_data = ["2024-01-01", "2024-02-01", "2024-03-01", "2024-04-01", "2024-05-01"]

plt.plot(y_data, marker = "o")
plt.ylim(bottom = 0, top = 10)
plt.xlim(left = 0, right = 6)

font_config = {"fontsize":22, "fontweight": "bold"}
plt.xticks([0, 2, 4, 6], ["A", "B", "C", "D"], **font_config)
plt.show()
```
