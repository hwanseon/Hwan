#### [Matplotlib 활용한 시각화]
- matplotlib.pyplot 모듈을 사용
- 다양한 그래프 관련 함수들 & 클래스들 존재

```python
# [1] 모듈 로딩
# -    패키지명.모듈명
import matplotlib.pyplot as plt
```

```python
# [2] 그래프 그리기 : 선(Line) Graph
# 기본 : plot(y축 데이터)
plt.plot([1, 2, 3, 4]) # 값이 1개 들어가면  Y 값
```

```python
plt.plot([1, 2, 3, 4])
plt.xlabel("DATA")
plt.ylabel("RESULT") 
plt.show()
```

```python
# [2] Graph 그리기 - Line(선) Graph
# plot(x축 데이터, y축 데이터)
plt.plot([10, 20, 30], [5, 10, 15])
plt.show()
```

```python
# [2] Graph 그리기 - Line(선) Graph
# plot(x축 데이터, y축 데이터, fmf포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10, 20, 30], [5, 10, 15],"r--") # 그래프의 색상과 선 모양을 표시함
plt.show()
```

```python
# [2] Graph 그리기 - Line(선) Graph
# plot(x축 데이터, y축 데이터, fmf포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10, 20, 30], [5, 10, 15],"ro--") # o를 지정해주면 각 점이 만나는 곳을 표시해줌
plt.show()
```

```python
# [2] Graph 그리기 - Line(선) Graph
# plot(x축 데이터, y축 데이터, fmf포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10, 20, 30], [5, 10, 15],"r^--") # ^를 하면 각 점이 만나는 지점이 위로 가는 삼각형으로 표시됨
```

```python
# [2] Graph 그리기 - Line(선) Graph
# plot(x축 데이터, y축 데이터, fmf포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10, 20, 30], [5, 10, 15],"b<--") # < or >를 지정해주면 각 점이 만나는 곳을 삼각형으로 표시해줌
plt.show()
```

```python
# [2] Graph 그리기 - Line(선) Graph
# plot(x축 데이터, y축 데이터, fmf포멧스트링)
# fmf 포맷스트링 : 그래프 색상, 두께, 스타일 등등 지정
plt.plot([10, 20, 30], [5, 10, 15],"ko") # 블루, 블랙 b로 겹치기 때문에 블랙은 K
# 포맷스트링에 선 모양의 대한 정보를 안주면 선 없이 점만 찍어줌 
plt.show()
```

```python
# [2] Graph 그리기 - Line(선) Graph
# plot(x축 데이터, y축 데이터, line2D 속성으로 설정)
# color = "red", marker = "o", linestyle = "dashed",
#           markersize = 10, linewidth = 4
# markeredgecolor => mec = "blue"
# markeredgewidth => mew = 5

plt.plot([10, 20, 30], [5, 10, 15], color = "red", marker = "o", linestyle = "dashed",
         markersize = 10, linewidth = 4, mec = 'k', mew = 5) 
plt.show()
```

```python
# [2] Graph 그리기 - Line(선) Graph
# 선 2개 그리기
plt.plot([10, 20, 30], [15, 25, 35], 'bo-') # 그래프를 그리기 위한 데이터
plt.plot([10, 20, 30], [18, 29, 25], 'ro--')
plt.plot([10, 20, 30], [25, 25, 25], 'k^-')
plt.show()
```

```python
# 위 선 3개를 하나로 합치는 코드
plt.plot([10, 20, 30], [15, 25, 35], 'bo-', [10, 20, 30], [18, 29, 25], 'ro--', [10, 20, 30], [25, 25, 25], 'k^-')
```

```python
# [3] 그래프의 부가적인 기능 추가
# - 그래프 제목
# - x, y축 라벨
# - 여러개의 그레프인 경우 범례 => 그래프의 label 속성값을 범례에서 사용
plt.plot([10, 20, 30], [15, 25, 20], 'bo--')
plt.title('[Metplotlib Line2D]')
plt.xlabel("DATA")
plt.ylabel("RESULT")
plt.legend("values") # 그래프의 label 속성과 연결됨 plt.plot([10, 20, 30], [15, 25, 20], 'bo--', label = "V")
# 라고 주면 plt.legend() 로 입력하면 됨
plt.show()
```

```python
# [3] 그래프의 부가적인 기능 추가
# - 그래프 제목
# - x, y축 라벨
# - 여러개의 그레프인 경우 범례 => 그래프의 label 속성값을 범례에서 사용
plt.plot([10, 20, 30], [15, 25, 20], 'bo--', label = "V")
plt.plot([10, 20, 30], [5, 30, 28], 'ro--', label = "2")
plt.title('[Metplotlib Line2D]')
plt.xlabel("DATA")
plt.ylabel("RESULT")
plt.legend() # 범례의 위치가 자기가 들어갈 수 있는 곳으로 컴퓨터가 알아서 옮겨감
plt.show()
```

```python
# [3] 그래프의 부가적인 기능 추가
# - 그래프 제목
# - x, y축 라벨
# - 여러개의 그레프인 경우 범례 => 그래프의 label 속성값을 범례에서 사용
#                                LOC 속성으로 위치 지정 가능 [기:best]
plt.plot([10, 20, 30], [15, 25, 20], 'bo--', label = "V")
plt.plot([10, 20, 30], [5, 30, 28], 'ro--', label = "2")
plt.title('[Metplotlib Line2D]')
plt.xlabel("DATA")
plt.ylabel("RESULT")
plt.legend(loc = "upper center") # lowerr center 등등 많음
plt.show()
```
