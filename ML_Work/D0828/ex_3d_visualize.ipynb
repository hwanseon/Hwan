#### [3D 시각화 예시]

```python
# 모듈로딩
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 데이터 생성
x = np.linspace(-10, 10, 11)
print(f"[x] {x.shape} \n{x}\n")

X = np.tile(x, (11, 1)) # (11,1)가 shape
# 즉 행을 11개 만드는 것
print(f"[X] {X.shape} \n{X}\n")

Y = np.transpose(X) 
print(f"[Y] {Y.shape} \n{Y}\n")

Z = np.random.rand(11, 11) # ramd: 0.0 ~ 1.0 사이의 값
print(f"[Z] {Z.shape} \n{Z}\n")
```

```python
# 시각화 진행
fig = plt.figure(figsize = (10, 8))

# 3차원 창 설정 => projection='3d' 
ax = fig.add_subplot(111, projection='3d') # projection : 투영 -> 여기서는 3D로 하겠다는 의미

# 평면 생성 : 2차원 어레이를 3차원 Surface로 표현
ax.plot_surface(X, Y, Z, color='yellow') # 표면 생성, 단순히 면을 하나 그리는 것
ax.set_zlim(-5, 5) # z축의 limit 범위 지정

# x, y, z축 레이블 지정
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.tight_layout()
plt.show()
```

- [예시2] 

```python
# 모듈로딩
from mpl_toolkits.mplot3d import Axes3D # 3차원 그래프를 그릴때 사용하는 모듈
import matplotlib.pyplot as plt # 2차원 그래프를 그릴때 사용하는 모듈
from matplotlib import cm
import numpy as np
```

```python
## 임의의 데이터 
x = [0, 0, 1, 1]
y = [0, 1, 0, 1]
z = [-1, -1, -1, -1]

## 평면 좌표 데이터 생성
# - 좌표 범위 생성
X_data = np.arange(-0.5, 1.5, 0.1)  # python은 실수를 사용하지 못하지만 np는 가능
Y_data = np.arange(-0.5, 1.5, 0.1)

# - 행렬 ==> 좌표값 변환
X, Y = np.meshgrid(X_data, Y_data) # meshgrid 행렬을 좌표값으로 변환
Z = (float(-5/7) * Y) + (float(-4/7) * X)
```

```python
## 시각화 
fig = plt.figure()
# projection : 객체가 그래프에 투영될 방법 설정
#              None, 'mollweide', 'polar', 'rectilinear', str
ax = fig.add_subplot(111, projection='3d') # -> 3D로 표시


# 데이터 출력
# 좌표평면 위에 점을 찍는 코드
ax.plot(x, y, z, 
        linestyle="none", 
        marker="o", mfc="none", markeredgecolor="red") # mfc: marker 안에 색깔 즉, 여기서는 마커 안에 색깔에 넣지말라는 의미

# 데이터 평면 출력
# stride, cstride : row와 column에서 데이터에서 어느정도 graph 그리는데 사용할지 설정
# alpha : 그래프의 투명도  
# cmap : 색상 설정 color map

# 좌표평면 위에 면으로 나타내는 코드
# 점들을 포함하는 ? 면 or 점들의 방향 경향을 따져 생성되는 면 
ax.plot_surface(X, Y, Z, 
                rstride=4, cstride=4, 
                alpha=0.4, cmap=cm.jet) 

plt.show()
```
