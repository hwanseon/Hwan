#### 피쳐 전처리
- 수치형 피쳐 전처리 ==> 스케일링
- 범주형 피쳐 전처리 ==> 인코딩

```python
# 모듈로딩
from sklearn.preprocessing import LabelEncoder
```

```python
items = ["TV", "냉장고", "선풍기", "에어컨"]
```

```python
# 라벨/정수 인코더 인스턴스 생성
lencoder = LabelEncoder()
```

```python
# 라벨/정수 인코더의 범위 및 맵핑 작업
lencoder.fit(items)
```

```python
# 라벨/정수값으로 변환
lencoder.transform(items)
```

### One-Hot Encoding

```python
### One-Hot Encoding 원핫인코더 모듈 로딩
from sklearn.preprocessing import OneHotEncoder
```

```python
# 인코더 인스턴스 생성
ohEncoder = OneHotEncoder()
```

```python
# 원핫 인코더의 범위 및 맨핍 작업
# fit 안에 2D 값을 저장해야함
ohEncoder.fit([items])
```

```python
# 원핫 인코딩으로 변환
ohEncoder.transform([items])
# 결과값을 보여주지 않음 => 풀어서 봐야함
```

```python
oh_items = ohEncoder.transform([items])
type(oh_items)
```

```python
oh_items.toarray()
```

```python
## numpy로 2차원 만들어서 값 구하기
```

```python
import numpy as np
```

```python
arr_items = np.array(items) # 1차원
arr_items = np.array(items).reshape(-1, 1) # 2차원

print(arr_items)
```

```python
ohEncoder.fit(arr_items)
```

```python
oh_items = ohEncoder.transform(arr_items)
type(oh_items)
```

```python
oh_items.toarray()
```

#### 정수 인코딩 ==> 범주형 feature 

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
# 타이타닉 자료를 예시
data = [["male", "C", "child"], ["female", "S", "child"]] 
```

```python
# 정수인코더 인스턴스 생성
odEncoder = OrdinalEncoder()

# 정수 인코딩 범위 및 맵핑 진행
odEncoder.fit(data) # fit(2D)
```

```python
# 범주 피처 -> 정수 피쳐 변환
odEncoder.transform(data)

# man, female 중에서 알파벳 순으로 하면 f가 먼저 -> female 0
# c, s증에 c가 먼저 -> c = 0
# child, child는 동일 값 -> 0
```
