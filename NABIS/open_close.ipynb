```python
region = pd.read_csv(r"C:\Hwan\NABIS\data\전국별 병원 개폐업 수.csv")
regionDF = pd.DataFrame(region)
regionDF
```

```python
regionDF.drop(index=[0], inplace=True)

regionDF = regionDF.reset_index(drop=True)
regionDF.columns = ["구분", "2019 개업", "2019 폐업", "2020 개업", "2020 폐업", "2021 개업", "2021 폐업", "2022 개업", "2022 폐업", "2023 개업", "2023 폐업"]
regionDF.to_csv("./지역별 병원 개폐업.csv")
regionDF
```

```python
# 데이터프레임의 수치 데이터 부분을 int로 변환
regionDF.iloc[:, 1:] = regionDF.iloc[:, 1:].astype(int)

# '수도권' 행을 추가 (서울 + 경기)
seoul_gyunggi_sum = regionDF.loc[regionDF['구분'] == '서울'].iloc[:, 1:].values + regionDF.loc[regionDF['구분'] == '경기'].iloc[:, 1:].values

# '수도권' 데이터프레임 생성
seoul_gyunggi_df = pd.DataFrame(seoul_gyunggi_sum, columns=regionDF.columns[1:])
seoul_gyunggi_df.insert(0, '구분', '수도권')

# 나머지 지역을 '지방'으로 묶기 (서울과 경기 제외)
others = regionDF[~regionDF['구분'].isin(['서울', '경기'])].set_index('구분')

# '지방' 행을 추가 (나머지 지역 합산)
province_sum = others.sum().values
province_df = pd.DataFrame([province_sum], columns=others.columns)
province_df.insert(0, '구분', '지방')

# 수도권과 지방을 합쳐서 최종 데이터프레임 생성
final_df = pd.concat([seoul_gyunggi_df, province_df], ignore_index=True)

# 결과 확인
print(final_df)
```

```python
final_df
```

```python
final_df.columns = ["구분", "2019 개업", "2019 폐업", "2020 개업", "2020 폐업", "2021 개업", "2021 폐업", "2022 개업", "2022 폐업", "2023 개업", "2023 폐업"]
# years = ["2019 개업", "2019 폐업", "2020 개업", "2020 폐업", "2021 개업", "2021 폐업", "2022 개업", "2022 폐업", "2023 개업", "2023 폐업"]

final_df.to_csv("./수도권 지방 개폐업.csv")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 시각화를 위해 '구분'을 제외한 나머지 수치 데이터 선택
plot_data = final_df.set_index('구분')

# 연도별로 개업 및 폐업에 대한 데이터를 나눠서 사용
years = ["2019 개업", "2019 폐업", "2020 개업", "2020 폐업", "2021 개업", "2021 폐업", "2022 개업", "2022 폐업", "2023 개업", "2023 폐업"]

# X축 위치 설정
x = np.arange(len(years))

# 그래프 크기 설정
plt.figure(figsize=(12, 6))

# 막대 너비 설정
width = 0.35

# 수도권 데이터 막대 그래프
plt.bar(x - width/2, plot_data.loc['수도권'], width=width, label='수도권')

# 지방 데이터 막대 그래프
plt.bar(x + width/2, plot_data.loc['지방'], width=width, label='지방')

# 그래프 꾸미기
plt.title('수도권과 지방의 연도별 개업 및 폐업 현황')
plt.xlabel('연도 및 상태')
plt.ylabel('개업/폐업 수')
plt.xticks(x, years, rotation=45)
plt.legend()

# 그래프 보여주기
plt.tight_layout()
plt.show()
```
