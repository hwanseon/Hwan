```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import pandas as pd

# ChromeDriver 경로 설정 (본인의 경로에 맞게 수정)
chrome_driver_path = r"C:\Users\hwans\Downloads\chromedriver-win64\chromedriver-win64\chromedriver.exe"

# Chrome 옵션 설정 (헤드리스 모드를 끔)
chrome_options = Options()
chrome_options.add_argument('--disable-gpu')

# ChromeDriver 서비스를 설정하여 Chrome 실행
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# wtable 사이트 열기
driver.get('https://wtable.co.kr/recipes')

# 페이지 로딩 대기
time.sleep(3)

# WebDriverWait으로 특정 요소가 로딩될 때까지 대기
wait = WebDriverWait(driver, 30)
wait.until("div", class_='RecipeListstyle__RecipeList-sc-1s9b4ly-20 eKgfiL')

# 페이지 소스를 가져와 BeautifulSoup으로 파싱
soup = BeautifulSoup(driver.page_source, 'html.parser')

# 밑반찬 카테고리의 URL을 크롤링
recipe_links = []
for link in soup.find_all('a', class_='RecipeListstyle__RecipeList-sc-1s9b4ly-20 eKgfiL'):
    recipe_links.append('https://wtable.co.kr' + link['href'])

# 각 레시피 상세 페이지에 접근하여 음식명, 재료, 이미지 URL 크롤링
recipes_data = []

for recipe_link in recipe_links:
    driver.get(recipe_link)
    time.sleep(3)
    
    # 상세 페이지 소스를 파싱
    recipe_soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    # 음식명 추출
    recipe_name = recipe_soup.find('h2', class_='RecipeDetailstyle__Title-q7sykd-4').text.strip()
    
    # 재료 목록 추출 (클래스 이름이 바뀔 수 있음, 실제 구조 확인 필요)
    ingredient_sections = recipe_soup.find_all('div', class_='Text__Description02-sc-1qy6bx2-0')
    
    ingredients = []
    for section in ingredient_sections:
        ingredient_name = section.find_all('div')[0].text.strip()  # 재료 이름
        ingredient_quantity = section.find_all('div')[1].text.strip()  # 재료 양
        ingredients.append(f"{ingredient_name}: {ingredient_quantity}")
    
    # 이미지 URL 추출
    image_tag = recipe_soup.find('img', class_='RecipeDetailstyle__Image-q7sykd-5')  # 이미지 태그 클래스 확인 필요
    image_url = image_tag['src'] if image_tag else '이미지 없음'

    # 레시피 데이터 저장
    recipes_data.append({
        'name': recipe_name,
        'ingredients': ', '.join(ingredients),
        'image_url': image_url
    })

# 크롬 드라이버 종료
driver.quit()

# pandas DataFrame으로 변환
df = pd.DataFrame(recipes_data)

# DataFrame을 CSV 파일로 저장
df.to_csv('wtable_recipe.csv', index=False)

# 데이터프레임을 출력해 확인
print(df.head())
```
