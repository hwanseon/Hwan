```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# 크롬 옵션 설정
chrome_options = Options()
chrome_options.add_experimental_option("detach", True)  # 브라우저 꺼짐 방지
chrome_options.add_experimental_option("useAutomationExtension", False)
chrome_options.add_experimental_option("excludeSwitches", ["enable-logging"])

# Chrome driver Manager를 통해 크롬 드라이버 자동 설치
service = Service(executable_path=ChromeDriverManager().install())

# Selenium WebDriver 실행
driver = webdriver.Chrome(service=service, options=chrome_options)

# 목표 페이지 URL
url = 'https://wtable.co.kr/recipes'
driver.get(url)

# 페이지가 완전히 로드될 때까지 기다림 (최대 10초)
wait = WebDriverWait(driver, 10)

# "일식" 카테고리 버튼 클릭
try:
    category_button = wait.until(EC.element_to_be_clickable((By.XPATH, "//li[contains(text(), '일식')]")))
    category_button.click()
    print("일식 카테고리 클릭 성공")
    
    # 일식 카테고리가 로드될 때까지 대기 (특정 요소로 대기할 수 있음)
    wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div[class^='RecipeListstyle__RecipeList']")))
    print("일식 카테고리 페이지 로드 완료")
except Exception as e:
    print(f"일식 카테고리 클릭 실패: {e}")

# 스크롤
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  
    new_height = driver.execute_script("return document.body.scrollHeight")

    if new_height == last_height:
        break
    last_height = new_height

# 각 레시피 클릭 후 상세 정보 크롤링
recipes = []

try:
    recipe_cards = driver.find_elements(By.CSS_SELECTOR, "div[class^='RecipeListstyle__RecipeList'] a")
    print(f"레시피 카드 개수: {len(recipe_cards)}")

    for card in recipe_cards:
        card.click()  # 레시피 카드 클릭
        time.sleep(2)  # 페이지 로드 대기

        try:
            # 소제목 가져오기 (XPATH 방식으로 대체)
            subtitle = driver.find_element(By.XPATH, "//h3[contains(@class, 'SubTitle')]").text
            
            # 음식명 가져오기 (XPATH 방식으로 대체)
            food_name = driver.find_element(By.XPATH, "//h2[contains(@class, 'Title')]").text
            
            # 재료 가져오기 (XPATH 방식으로 대체)
            ingredients = driver.find_element(By.XPATH, "//div[contains(@class, 'Text__Description02')]").text
            
            # 현재 페이지 URL 저장
            recipe_url = driver.current_url
            
            # 링크 저장
            recipe_data = {
                '음식명': food_name,
                '소제목': subtitle,
                '재료': ingredients,
                '링크': recipe_url
            }
            
            recipes.append(recipe_data)

        except Exception as e:
            print(f"Error while processing recipe: {e}")
        
        driver.back()  # 뒤로 가기 (레시피 목록으로 돌아가기)
        time.sleep(2)  # 페이지 로드 대기

except Exception as e:
    print(f"레시피 크롤링 실패: {e}")

# 크롤링한 데이터를 DataFrame으로 변환
if recipes:
    df = pd.DataFrame(recipes)

    # DataFrame을 CSV 파일로 저장 (VSCode에서 확인할 수 있도록)
    df.to_csv('japnRecipes.csv', index=False, encoding='utf-8-sig')

    print("CSV 파일로 저장되었습니다. 'recipes.csv' 파일을 확인하세요.")
else:
    print("크롤링된 데이터가 없습니다. 확인이 필요합니다.")

# 브라우저 종료
driver.quit()
```

```python
# 구두점 제거
def remove_punctuation(text):
    punc = string.punctuation
    # 문자열인 경우에만 구두점을 제거
    if isinstance(text, str):
        return ''.join([char for char in text if char not in punc])
    else:
        return text
```

```python
import numpy as np
from collections import Counter

# 토큰화 함수
def tokenize(text):
    if isinstance(text, str):
        return text.split()  
    else:
        return []

TokenList = JDF['소제목'].apply(tokenize).tolist()

# 모든 단어에 대한 단어사전(단어 -> 정수 매핑) 만들기
counter = Counter()
for tokens in TokenList:
    counter.update(tokens)

vocab_size = 100
most_common_words = counter.most_common(vocab_size)
word_to_index = {word: i + 1 for i, (word, _) in enumerate(most_common_words)} 
# 문장을 정수 인코딩
encoded_sentences = []
for tokens in TokenList:
    encoded_sentence = [word_to_index.get(word, 0) for word in tokens]  
    encoded_sentences.append(encoded_sentence)

# 결과 확인
print("정수 인코딩 결과 (상위 5개 문장):")
print(encoded_sentences[:5])
```

```python
maxlen = 10

# 패딩을 추가하는 함수
def pad_sequence(sequence, maxlen):
    if len(sequence) < maxlen:
        return sequence + [0] * (maxlen - len(sequence))  # 짧으면 0을 뒤에 채움
    else:
        return sequence[:maxlen]  # 길면 자름

# 모든 문장에 패딩 적용
padded_sentences = np.array([pad_sequence(sentence, maxlen) for sentence in encoded_sentences])

# 결과 확인
print("패딩된 결과 (상위 5개 문장):")
print(padded_sentences[:5])
```

- DataLoader

```python
import torch
from torch.utils.data import Dataset, DataLoader

class TextDataset(Dataset):
    def __init__(self, data, labels):
        self.data = data 
        self.labels = labels 
    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return torch.tensor(self.data[idx], dtype=torch.long), torch.tensor(self.labels[idx], dtype=torch.long)
```

```python
# 필요한 라이브러리 임포트
import torch
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.optim as optim

# TextDataset 클래스는 이미 정의되어 있으니, 이를 이용해서 학습에 필요한 데이터로더를 생성합니다.

# 데이터 준비 과정에서 정수 인코딩된 `padded_sentences`와 레이블 `y_train`, `y_test`를 이용
train_dataset = TextDataset(padded_sentences[:len(X_train)], y_train.values)
test_dataset = TextDataset(padded_sentences[len(X_train):], y_test.values)

# 데이터로더 생성
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# 간단한 모델 정의 (예시로 RNN 사용)
class SimpleRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim):
        super(SimpleRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        embedded = self.embedding(x)
        output, hidden = self.rnn(embedded)
        out = self.fc(hidden.squeeze(0))
        return out
    
from sklearn.preprocessing import LabelEncoder

# LabelEncoder를 전체 데이터에 대해 fit
label_encoder = LabelEncoder()

# 학습 데이터와 테스트 데이터를 합친 후 fit
all_labels = pd.concat([y_train, y_test], axis=0)
label_encoder.fit(all_labels)

# 각각의 라벨을 정수로 변환
y_train_encoded = label_encoder.transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# TextDataset에 인코딩된 라벨을 사용
train_dataset = TextDataset(padded_sentences[:len(X_train)], y_train_encoded)
test_dataset = TextDataset(padded_sentences[len(X_train):], y_test_encoded)

# 데이터로더 생성
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```
