```python
import string

string.punctuation
```

```python
# 구두점 제거 함수
def remove_punctuation(text):
    punc = string.punctuation
    # 문자열인 경우에만 구두점을 제거
    if isinstance(text, str):
        return ''.join([char for char in text if char not in punc])
    else:
        return text
```

```python
# Train 데이터에서 구두점 제거
X_train['소제목'] = X_train['소제목'].apply(remove_punctuation)

# Test 데이터에서 구두점 제거
X_test['소제목'] = X_test['소제목'].apply(remove_punctuation)
```

```python
import numpy as np
from collections import Counter

# 토큰화 함수
def tokenize(text):
    if isinstance(text, str):
        return text.split()  
    else:
        return []

TokenList = JDF['소제목'].apply(tokenize).tolist()

# 모든 단어에 대한 단어사전(단어 -> 정수 매핑) 만들기
counter = Counter()
for tokens in TokenList:
    counter.update(tokens)

vocab_size = 100
most_common_words = counter.most_common(vocab_size)
word_to_index = {word: i + 1 for i, (word, _) in enumerate(most_common_words)} 
# 문장을 정수 인코딩
encoded_sentences = []
for tokens in TokenList:
    encoded_sentence = [word_to_index.get(word, 0) for word in tokens]  
    encoded_sentences.append(encoded_sentence)

# 결과 확인
print("정수 인코딩 결과 (상위 5개 문장):")
print(encoded_sentences[:5])
```

```python
maxlen = 10

# 패딩을 추가하는 함수
def pad_sequence(sequence, maxlen):
    if len(sequence) < maxlen:
        return sequence + [0] * (maxlen - len(sequence))  # 짧으면 0을 뒤에 채움
    else:
        return sequence[:maxlen]  # 길면 자름

# 모든 문장에 패딩 적용
padded_sentences = np.array([pad_sequence(sentence, maxlen) for sentence in encoded_sentences])

# 결과 확인
print("패딩된 결과 (상위 5개 문장):")
print(padded_sentences[:5])
```

- DataLoader

```python
import torch
from torch.utils.data import Dataset, DataLoader

class TextDataset(Dataset):
    def __init__(self, data, labels):
        self.data = data 
        self.labels = labels 
    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return torch.tensor(self.data[idx], dtype=torch.long), torch.tensor(self.labels[idx], dtype=torch.long)
```

```python
# 필요한 라이브러리 임포트
import torch
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.optim as optim

# TextDataset 클래스는 이미 정의되어 있으니, 이를 이용해서 학습에 필요한 데이터로더를 생성합니다.

# 데이터 준비 과정에서 정수 인코딩된 `padded_sentences`와 레이블 `y_train`, `y_test`를 이용
train_dataset = TextDataset(padded_sentences[:len(X_train)], y_train.values)
test_dataset = TextDataset(padded_sentences[len(X_train):], y_test.values)

# 데이터로더 생성
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# 간단한 모델 정의 (예시로 RNN 사용)
class SimpleRNN(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, output_dim):
        super(SimpleRNN, self).__init__()
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        embedded = self.embedding(x)
        output, hidden = self.rnn(embedded)
        out = self.fc(hidden.squeeze(0))
        return out
    
from sklearn.preprocessing import LabelEncoder

# LabelEncoder를 전체 데이터에 대해 fit
label_encoder = LabelEncoder()

# 학습 데이터와 테스트 데이터를 합친 후 fit
all_labels = pd.concat([y_train, y_test], axis=0)
label_encoder.fit(all_labels)

# 각각의 라벨을 정수로 변환
y_train_encoded = label_encoder.transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# TextDataset에 인코딩된 라벨을 사용
train_dataset = TextDataset(padded_sentences[:len(X_train)], y_train_encoded)
test_dataset = TextDataset(padded_sentences[len(X_train):], y_test_encoded)

# 데이터로더 생성
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
# vocab_size 정의
vocab_size = len(set([word for sentence in padded_sentences for word in sentence]))  # 고유 단어의 수

embedding_dim = 100  # 단어 임베딩의 차원
hidden_dim = 128  # RNN의 은닉층 차원
num_classes = len(label_encoder.classes_)  # 출력 클래스의 수 (라벨의 수)

# 모델 설정
model = SimpleRNN(vocab_size, embedding_dim, hidden_dim, num_classes)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 학습 루프
n_epochs = 280
for epoch in range(n_epochs):
    model.train()
    total_loss = 0
    correct_train = 0
    total_train = 0
    
    for inputs, labels in train_loader:
        optimizer.zero_grad()
        outputs = model(inputs)
        
        # 손실(loss) 계산
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

        # 학습 데이터의 예측 정확도 계산
        _, predicted = torch.max(outputs, 1)
        total_train += labels.size(0)
        correct_train += (predicted == labels).sum().item()

    avg_loss = total_loss / len(train_loader)
    train_accuracy = correct_train / total_train

    # 테스트 데이터로 평가
    model.eval()
    correct_test = 0
    total_test = 0
    with torch.no_grad():
        for inputs, labels in test_loader:
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            total_test += labels.size(0)
            correct_test += (predicted == labels).sum().item()

    test_accuracy = correct_test / total_test

    # 각 에포크의 loss와 accuracy 출력
    print(f'Epoch {epoch+1}/{n_epochs}, Loss: {avg_loss:.4f}, '
          f'Train Accuracy: {train_accuracy:.4f}, Test Accuracy: {test_accuracy:.4f}')

# 모델을 pth 파일로 저장
torch.save(model.state_dict(), 'model.pth')
```
