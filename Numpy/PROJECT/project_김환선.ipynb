```python
X = migraine[['Age', 'Photophobia', 'Vomit']].values

# 편두통이 있는 경우 : 1, 없는 경우 : 0
y = (migraine['Location'] != 0).astype(int).values
```

```python
# 데이터 표준화 

X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)

print(X_mean, X_std)
X = (X - X_mean) / X_std
```

```python
# 로지스틱 회귀 함수 구현

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def predict(X, weights):
    z = np.dot(X, weights)
    return sigmoid(z)
```

```python
# 모델 학습 (경사 하강법 사용)

def train(X, y, learning_rate=0.01, iterations=1000):
    weights = np.zeros(X.shape[1])
    for i in range(iterations):
        predictions = predict(X, weights)
        gradient = np.dot(X.T, (predictions - y)) / y.size
        weights -= learning_rate * gradient
    return weights
```

```python
# 모델 훈련

weights = train(X, y)
```

```python
from matplotlib.lines import Line2D
```

```python
# 예측

predictions = predict(X, weights)
predicted_classes = [1 if p >= 0.5 else 0 for p in predictions]

# 예측 결과 시각화

plt.scatter(migraine['Age'], predictions, c=predicted_classes, cmap='bwr', edgecolors='black')
plt.title('[나이에 따른 편두통 발생 확률]')
plt.xlabel('나이')
plt.ylabel('편두통 발생 확률')

legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='편두통 없음', markerfacecolor='blue', markersize=10, markeredgecolor='black'),
    Line2D([0], [0], marker='o', color='w', label='편두통 발생', markerfacecolor='red', markersize=10, markeredgecolor='black')
]
plt.legend(handles=legend_elements, loc='best')
plt.show()
```
