#### Day 01 정리
- 텍스트 데이터 ==> 말뭉치 / 코퍼스
- 형태소 분리 ==> 토큰 (단어, 문장, 문자(바이트))
- 불필요한 토큰 제거 ==> 구두점 + 추가
 
            -      ==> 빈도는 높은데 의미가 없는 토큰(인칭대명사, 관사, 조사, ...)
             
            -      ==> 빈도가 낮은 단어들도 검사를 해야함 -> 중요하지 않은 토큰 제거
             
            -      ==> 길이가 짧은 토큰 
             
            - =====> 이 모든것을 정제 작업이라고 함

- 어간추출, 표제어 제거
- 중복 토큰, 의미가 동일한 토큰 통합 제거 => 동일 부분 추출 (동사: 동일 부분 추출)

- 텍스트 ==> 수치화 (수치화를 하려면 기준이 있어야함)

- 인코딩 : 우리가 사용하는 언어 -> 기계어
- 디코딩 : 기계어 -> 우리가 사용하는 언어
